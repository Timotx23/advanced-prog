## Threads and Processes

When you want different parts of your program running concurrently (i.e., in parallel), you can choose between two ways of implementing this behavior: **threading** and **multiprocessing**. 

With threading, the program implements multiple functionalities which are "interleaved," i.e., the program switches between the functionalities. The functionalities have shared memory and shared processing power. The Global Interpreter Lock (GIL) ensures that only one thread is executed at a time (even if multiple processors would be available).

With multiprocessing, the program implements multiple functionalities, each of which uses their own processor. Therefore they run truly "in parallel." Each process has its own memory. You may view this as each process being its own program.

## Threading

Python provides the <code>threading</code> library to implement threading: https://docs.python.org/3/library/threading.html

An implementation of a thread can be done by calling the object constructor of a thread with a function as a keyword argument.

Create a thread, after `import threading`, with the command:<br> 
`<thread> = threading.Thread( target=<function>, name=<name>, args=<arguments> )`<br>
where `<thread>` is a variable, `<function>` is a function (mandatory), `<name>` is a string (optional), and `<args>` is a list or set of arguments given to `<function>` (optional).

Start the thread with the command: `<thread>.start()`

Make sure that the thread will finish before the program finishes with the command: `<thread>.join()`

Example:

```python
# NB1_01.py

import threading
import time

def sayMessage( message, reps ): 
    for _ in range( reps ):      # `_` is commonly used as a throwaway variable, a placeholder in loops
        print( message )
        time.sleep(1)

t1 = threading.Thread( target=sayMessage, args = ("Hello, world!", 10) )
t1.start() 
print('We can run other instructions while the thread is alive.')
time.sleep(3) #pause the main program
print('The thread keeps running while the main program sleeps.')
t1.join() 
print('end')
```

Example which creates multiple threads (note that the main program itself is also a thread, so there are three threads here):

```python
# NB1_02.py

import threading
import time

def sayMessage( message, reps ): 
    for _ in range( reps ):      # `_` is commonly used as a throwaway variable, a placeholder in loops
        print( message )
        time.sleep(1)
        
t1 = threading.Thread( target=sayMessage, args=("Hello, world!", 10) ) 
t2 = threading.Thread( target=sayMessage, args = ("How are you today?", 5) ) 

t1.start() 
t2.start() 
t1.join() 
t2.join() 

print( "end" )
```

## Thread communication

Threads "live" within the process that creates them, *i.e.*, they have access to the memory of their parent process. This means that threads can see the variables that are in the scope of the thread creation call. You have to be careful with this, because you do not know when exactly a thread will read or change a global variable. If multiple threads work with the same variables, results may be unexpected.

## Thread communication with queues

From the `queue` module, use `Queue` (FIFO), `LifoQueue` (LIFO), or `PriorityQueue` (lowest out first) for thread-safe communication between threads. Call `put()` to put something in the queue, `get()` to get something out of the queue, `task_done()` to indicate that an element was taken from the queue (without calling it after `get()`, the queue is locked for further processing). Use `join()` for the queue at the end to ensure that all items have been processed.

Modify the example of *greet_once* such that it uses a Queue instead of a global variable, and thus actually only greets once. 

```python
# NB1_06.py

import threading, time
from queue import Queue

### Exercise 3 ###############################

import threading, time
SENTINAL = "END"
greeting = 'Hi'

def producer(input, item):
    for i in range(item):
        input.put(greeting)
    input.put(SENTINAL)
    


def greet_once(input, output): 
    while True:
        item = input.get()

        if item == SENTINAL:
            input.task_done()
            output.put(SENTINAL)
            break
        output.put(item)
        input.task_done()

    
def consumer(output, result):
    
    while True:
        item = output.get()
        if item == SENTINAL:
            output.task_done()
            break
        result.append(item)
        output.task_done()


threads = []
input_queue = Queue()
output_queue = Queue()




producer_thread = threading.Thread( target=producer, args=(input_queue,10 )) 
worker_thread = threading.Thread(target=greet_once, args=(input_queue, output_queue))
conumser_thread = threading.Thread(target=consumer, args=(output_queue, threads))

producer_thread.start(), worker_thread.start(), conumser_thread.start()
input_queue.join(), output_queue.join()
producer_thread.join(), worker_thread.join(), conumser_thread.join()

for b in threads:
    print(b)

print( "end" )


##############################################

```

### Producer-consumer example

Here is an example of a producer-thread which generates numbers, and two consumer-threads which process the numbers. They use two queues to communicate: the first stores the numbers generated by the producer, and the second stores the numbers which the consumers generate (which are actually the same as the producer's numbers). When the producer has finished its task, it places two SENTINEL objects (simply strings "END") in its queue, to communicate to the consumers that the queue is finished. There must be two SENTINELs, because there are two consumers. Finally, the main program prints the numbers from the second queue, which are the same as the producer generated, but may be in a different order, depending on how the consumers were interleaved.

Note that q2 is joined only after the program is finished with q2. If you would do it earlier, the program will get into an endless loop, waiting for more things to happen to q2.

```python
# NB1_07.py

import threading, time
from queue import Queue

SENTINEL = "END"

def producer( q, n ):
    for a in range( n+1 ):
        q.put(a)
    q.put( SENTINEL ); q.put( SENTINEL )
    
def consumer( q1, q2 ):
    while True:
        num = q1.get()
        time.sleep(0.1)
        q1.task_done()
        if num == SENTINEL:
            break
        q2.put( num )
    
q1 = Queue(); q2 = Queue()
cns1 = threading.Thread( target=consumer, args=(q1,q2) )
cns2 = threading.Thread( target=consumer, args=(q1,q2) )
prd = threading.Thread( target=producer, args=(q1,25) )

cns1.start(); cns2.start(); prd.start() 
q1.join(); cns1.join(); cns2.join(); prd.join()

while not q2.empty():
    num = q2.get()
    q2.task_done()
    print( num )

q2.join()

print('end')
```

## Multiprocessing

Python provides the <code>multiprocessing</code> library to implement multiprocessing: https://docs.python.org/3/library/multiprocessing.html

Please note the warning given earlier about multiprocessing under Windows: you may not see any outputs!

An implementation of a process can be done by calling the object constructor of a process with a function as a keyword argument.

Create a process, after `import multiprocessing`, with the command:<br> 
`<process> = multiprocessing.Process( target=<function>, name=<name>, args=<arguments> )`<br>
where `<process>` is a variable, `<function>` is a function (mandatory), `<name>` is a string (optional), and `<args>` is a list or set of arguments given to `<function>` (optional).

Start the process with the command: `<process>.start()`

Make sure that the process will finish before the program finishes with the command: `<process>.join()`

Note that the main program must also be a process, otherwise it will finish before the other processes have had a chance to run. The easiest way to achieve that is to place the main program under `if __name__ == "__main__":`

Example (this is a straightforward adaptation of the threading example given before; note that you may not see the outputs of the `sayMessage()` function, depending on your OS):

```python
# NB1_08.py

import multiprocessing
import time

def sayMessage( message, reps ): 
    for _ in range( reps ):      # `_` is commonly used as a throwaway variable, a placeholder in loops
        print( message )
        time.sleep(1)

if __name__ == "__main__":
    p1 = multiprocessing.Process( target=sayMessage, args = ("Hello, world!", 10) )
    p1.start() 
    print('We can run other instructions while the process is alive.')
    time.sleep(3) #pause the main program
    print('The process keeps running while the main program sleeps.')
    p1.join() 
    print('end')
```

## Communication between processes

We have introduced a `Queue` as a method for communication between `Thread`s. However, you may have noticed that the shared `Queue` was accessible to multiple `Thread`s because it was in their memory scope, as `Thread`s share the memory of the process that created them.

That's not the case for Processes. When you create a process, they get a **copy** of the memory of their parent. However, there is an implementation of a `Queue` that works for processes in the `multiprocessing` module. Its interface is very similar to the previously seen `Queue`, except that the `task_done()` and `join()` methods do not exist (and are not needed). Its implementation is such that each process still has its own queue, but every time a process makes a change to the queue, the changes are automatically transferred to the queues of all processes.

There are other ways to use shared memory with processes, all supported by the `multiprocessing` module. We will limit ourselves to queues here.

In the code below, to find a prime number bigger than the player's prime number, the computer simply starts with that number plus 1, and continues increasing it until it finds a prime.

Check the code and run it, making sure that you understand what is happening.


```python
# NB1_10.py

from math import sqrt

def is_prime( n ):
    if n < 2:
        return False
    i = 2
    while i <= sqrt( n ):
        if n % i == 0:
            return False
        i += 1
    return True

print( "2 is the lowest prime number" )
lastnum = 2
while True:
    num = int( input( "Give me a bigger prime number bigger than "+str( lastnum )+": " ) )
    if num <= lastnum:
        print( "That number is smaller than", lastnum )
        continue
    if not is_prime( num ):
        print( num, "is not prime. Goodbye!" )
        break
    lastnum = num+1
    while not is_prime( lastnum ):
        lastnum += 1
    print( lastnum, "is prime and bigger than", num )
```
