Use `filter()` to extract words longer than 5 characters from the list _words_.

**Hint:** Using `len()` may come in handy.

```python
words = ['apple', 'banana', 'cherry', 'date', 'fig']

## Your solution to HW-3 goes here:
long_words=list(

        filter(lambda x: x if len(x)>5 else "", words)
    )




###########################

print(list(long_words))  # Expected output: ['banana', 'cherry']
```

#### HW-9:

Implement the function  `add_all()` that takes an arbitrary number of arguments and sums them all, using `reduce()`.


```python
from functools import reduce

## Your solution to HW-10 goes here:
def add_all(*args):
    return reduce(lambda *args: sum(args), args)



###########################

print(add_all(1, 2, 3, 4)) # => 10
print(add_all(5, 10, 15))  # => 30
```

#### HW-10:

Implement the function `div_all()` that divides the first argument by all the remaining arguments sequentially. Use `reduce()` to implement it.

```python
## Your solution to HW-10 goes here:
def div_all(*args):
    return reduce(lambda *args: args[0]/(args[1]),args )



###########################

print(div_all(100, 2, 5))  # => 10.0
print(div_all(50, 5, 2))  # => 5.0
```

#### `combine()`

The function `combine()` is another auxiliary function that takes two functions and applies them in sequence. For example, `combine(triple, add_all)` creates a function that sums its arguments and then triples the result.


```python
def combine(f, g):
    return lambda *args, **kwargs: f(g(*args, **kwargs))

result = combine(triple, add_all)(1, 2, 3)  # => triple(add_all(1, 2, 3)) = triple(6) = 18
print(result)
```
